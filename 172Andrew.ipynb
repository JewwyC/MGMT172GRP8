```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

start1 = datetime.datetime(2020, 1, 1)
end1 = datetime.datetime(2020, 12, 31)

start2 = datetime.datetime(2019, 5, 20)
end2 = datetime.datetime(2024, 5, 20)

Apple2020 = yf.download("AAPL", start1, end1)
Apple5Year = yf.download("AAPL", start2, end2)


Apple2020["Close"].plot()

plt.xlabel('Date')
plt.ylabel('Closing Price (in USD)')
plt.grid(axis='y')
plt.show()

Apple5Year["Close"].plot()

plt.xlabel('Date')
plt.ylabel('Closing Price (in USD)')
plt.grid(axis='y')
plt.show()


```
