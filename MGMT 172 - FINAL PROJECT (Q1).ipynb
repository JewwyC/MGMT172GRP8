```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
start_date = '1980-01-01'
end_date = '2024-12-31'
```

##### retrieving Apple's stock data

```python
Apple = yf.download("AAPL", start_date, end_date)
```

```python
Apple["Company"]='Apple'
apple = yf.Ticker("AAPL")
```

```python
history = apple.history(start='1980-01-01', end='2024-12-31')
df = pd.DataFrame(history)
df.head (10)
```

```python
x = df.index
y = df['Close']
y
```

```python
def df_plot(data, x, y, title="", xlabel='Date', ylabel='Value', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
```

```python
stock_name= 'Apple'
title = (stock_name,'Historic Stock Performance')
df_plot(df , x , y , title=title,xlabel='Date', ylabel='Value',dpi=100)
```

##### retrieving Apple's product release dates/data

```python
tables= pd.read_html('https://en.wikipedia.org/wiki/List_of_Apple_products')
apple_df = pd.concat(tables[1:44])
selected_columns = apple_df[["Released", "Model","Family", "Discontinued"]]
selected_columns = selected_columns.dropna()
print(selected_columns)
```

#### product release dates vs. stock prices

```python
product_release_dates = apple_df['Released']
```

```python
import random
```

```python
plt.figure(figsize=(100, 6))
Apple['Close'].plot(title='Apple Stock Prices vs.Product Releases', xlabel='Date', ylabel='Price (USD)', grid=True)
product_names = apple_df['Model']

for release_date, product_name in zip(product_release_dates, product_names):
    release_date_formatted = pd.to_datetime(release_date, format='mixed')
    if release_date_formatted in Apple.index:
        random_color = [random.random() for _ in range(3)]
        plt.scatter(release_date_formatted, Apple.loc[release_date_formatted, 'Close'], color=random_color, label=product_name)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
Apple['Close'].plot(title='Apple Stock Prices vs.Product Releases', xlabel='Date', ylabel='Price (USD)', grid=True)
specific_release_dates = ['1984-01-24', '2007-06-29', '2010-04-03', '2014-09-09', '2015-04-24', '2017-11-03', '2018-09-12','2020-03-18', '2020-11-17', '2021-09-14', '2022-03-18', '2022-09-16','2023-11-07']
event_labels = ['Macintosh', 'iPhone', 'iPad', 'iPhone 6', 'Apple Watch', 'iPhone X', 'iPhone XS','iPad Pro', 'Apple M1 Chip', 'iPhone 13','iPhone SE', 'iPhone 14','Macbook Pro']

for release_date, label in zip(specific_release_dates, event_labels):
    release_date_formatted = pd.to_datetime(release_date)
    if release_date_formatted in Apple.index:
        random_color = [random.random() for _ in range(3)]
        plt.scatter(release_date_formatted, Apple.loc[release_date_formatted, 'Close'], color=random_color, label='Product Release')
        plt.text(release_date_formatted, Apple.loc[release_date_formatted, 'Close'], label, ha='right')

plt.legend(['Close'] + event_labels)
plt.show()
```

```python
lowest_price_index = Apple['Close'].idxmin()

closest_release_date = None
closest_product_name = None
min_distance = float('inf')

for release_date, product_name in zip(product_release_dates, product_names):
    release_date_formatted = pd.to_datetime(release_date)
    distance = abs((lowest_price_index - release_date_formatted).days)
    if distance < min_distance:
        min_distance = distance
        closest_release_date = release_date_formatted
        closest_product_name = product_name

print("Closest product release date to the lowest stock price of Apple:", closest_release_date)
print("Product released closest to the lowest stock price:", closest_product_name)
```

```python
stock_price_1982_10_01 = Apple.loc['1982-10-01', 'Close']
print("Apple stock price on October 1, 1982:", stock_price_1982_10_01)
```

```python
peak_index = Apple['Close'].idxmax()

closest_release_date = None
closest_product_name = None
min_distance = float('inf')

for release_date, product_name in zip(product_release_dates, product_names):
    release_date_formatted = pd.to_datetime(release_date)
    distance = abs((peak_index - release_date_formatted).days)
    if distance < min_distance:
        min_distance = distance
        closest_release_date = release_date_formatted
        closest_product_name = product_name

print("Closest product release date to the peak of stock price:", closest_release_date)
print("Product released closest to the peak:", closest_product_name)
```

#### scandals vs. stock prices

```python
plt.figure(figsize=(30, 6))
Apple['Close'].plot(title='Apple Stock Prices', xlabel='Date', ylabel='Price (USD)', grid=True)

specific_scandal_dates = ['2005', '2007', '2008', '2012', '2014', '2020', '2024']
scandal_labels = ['iPod/iTunes Antitrust', 'Apple and AT&T Mobility Antitrust', 'European Antitrust', 'eBook Price Fixing', 'High-Tech Employee Antitrust', 'Epic Games Lawsuit', 'Anticompetitive Litigation']

for date, label in zip(specific_scandal_dates, scandal_labels):
    random_color = [random.random() for _ in range(3)]
    plt.axvline(x=date, color=random_color, linestyle='--', label=label)
    plt.text(date, Apple['Close'].max()*0.6, label, rotation=90,va='top')

plt.legend()
plt.show()
```
