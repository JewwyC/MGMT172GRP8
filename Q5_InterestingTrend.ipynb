```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sns

start1 = datetime.datetime(2020, 1, 1)
end1 = datetime.datetime(2020, 12, 31)

start2 = datetime.datetime(2019, 5, 20)
end2 = datetime.datetime(2024, 5, 20)

start3 = datetime.datetime(2019, 1, 1)
end3 = datetime.datetime(2019, 12, 31)

Apple2020 = yf.download("AAPL", start1, end1)
Apple5Year = yf.download("AAPL", start2, end2)
Apple2019 = yf.download("AAPL", start3, end3)
Apple = yf.Ticker("AAPL")


Apple2020["Close"].plot()

plt.xlabel('Date')
plt.ylabel('Closing Price (in USD)')
plt.grid(axis='y')
plt.title('Apple Closing Prices (2020)')
plt.show()

Apple5Year["Close"].plot()

plt.xlabel('Date')
plt.ylabel('Closing Price (in USD)')
plt.grid(axis='y')
plt.title('Apple Closing Prices (5 Year)')
plt.show()







```

```python
Apple2020['SMA_50'] = Apple2020['Close'].rolling(window=50).mean()
Apple2020['SMA_200'] = Apple2020['Close'].rolling(window=200).mean()

plt.figure(figsize=(14, 7))
plt.plot(Apple2020['Close'], label='Close Price')
plt.plot(Apple2020['SMA_50'], label='50-Day SMA')
plt.plot(Apple2020['SMA_200'], label='200-Day SMA')
plt.title('Apple Stock Price with Simple Moving Averages (2020)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python
Apple5Year = yf.download('AAPL', start='2020-01-01', end='2024-01-01')

Apple5Year['SMA_50'] = Apple5Year['Close'].rolling(window=50).mean()
Apple5Year['SMA_200'] = Apple5Year['Close'].rolling(window=200).mean()

plt.figure(figsize=(14, 7))
plt.plot(Apple5Year['Close'], label='Close Price')
plt.plot(Apple5Year['SMA_50'], label='50-Day SMA')
plt.plot(Apple5Year['SMA_200'], label='200-Day SMA')
plt.title('Apple Stock Price with Simple Moving Averages (5 Year)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python
sp500 = yf.download('^GSPC', start='2020-01-01', end='2024-01-01')

# Calculate daily returns
Apple5Year['Return'] = Apple5Year['Close'].pct_change()
sp500['Return'] = sp500['Close'].pct_change()

# Compute correlation
correlation = Apple5Year['Return'].corr(sp500['Return'])
print(f'Correlation between Apple and S&P 500 returns: {correlation:.2f}')
```

```python
#Download other tech stock data 
stocks = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'VOO']
stockData = yf.download(stocks, start='2019-01-01', end='2024-01-01')['Close']

#Calculate the daily returns

returns = stockData.pct_change().dropna()

#Correlation matrix of returns
correlation_matrix = returns.corr()

#Plotting heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Apple versus Other Tech Stocks')
plt.show()

Microsoft = yf.Ticker("MSFT")
```
